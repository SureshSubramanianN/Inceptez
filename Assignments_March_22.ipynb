### 12.	Read the demand transaction file again and perform following operations for practice

a.	Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean

b.	One hot encode the columns Mapped_Sales_Type 

c.	Label binarize the column Day

d.	Label encode the column City


```python
from sklearn.impute import SimpleImputer
df = pd.read_csv(r'C:\Users\subramanian2\Downloads\Demand_txn_updated.csv')
df_col = df['Percent_Quantity_With_Discount']
df1 = pd.DataFrame(df_col)
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df1)
df_col=pd.DataFrame(imp.transform(df1))
df_col[:20]
```

```python
df_Mapped_Sales_Type = df['Mapped_Sales_Type']
df_dummies = pd.get_dummies(df_Mapped_Sales_Type, prefix=['D_'])
df_dummies
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
lb_output = lb.fit_transform(df['Day'])
print(type(lb_output))
pd.DataFrame(lb_output, columns=lb.classes_)
```
